## Question 1
The theoretical simple linear regression model relies on a predictor variable to determine whether it has a correlation association with the outcome variables. Slope and interecept cofeeicients are parameters that can provide information regarding the data. They combine to form a sample from normal distribution if the errors on average do not change the outcome variables drastically. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the regression model
beta_0 = 5      # Intercept
beta_1 = 2      # Slope
sigma = 1       # Standard deviation of the error term (normal distribution)

# Generate synthetic data
n = 100         # Number of data points
X = np.random.uniform(0, 10, size=n)  # Predictor variable X (random values between 0 and 10)

# Generate the error term (epsilon) from a normal distribution with mean 0 and std dev sigma
epsilon = np.random.normal(0, sigma, size=n)

# Generate the outcome variable Y: Y = beta_0 + beta_1 * X + epsilon
Y = beta_0 + beta_1 * X + epsilon

# Plot the data (Unfitted regression line)
plt.scatter(X, Y, color='blue', alpha=0.6, label="Data points")  # Data points
plt.plot(X, beta_0 + beta_1 * X, color='green', label="True regression line", linestyle='--')  # True regression line
plt.title("Unfitted Simple Linear Regression Model")
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.legend()
plt.show()
```

## Question Two

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the regression model
beta_0 = 5      # Intercept
beta_1 = 2      # Slope
sigma = 1       # Standard deviation of the error term (normal distribution)

# Generate synthetic data
n = 100         # Number of data points
X = np.random.uniform(0, 10, size=n)  # Predictor variable X (random values between 0 and 10)

# Generate the error term (epsilon) from a normal distribution with mean 0 and std dev sigma
epsilon = np.random.normal(0, sigma, size=n)

# Generate the outcome variable Y: Y = beta_0 + beta_1 * X + epsilon
Y = beta_0 + beta_1 * X + epsilon

# Create a pandas DataFrame to store the data
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a linear regression model using statsmodels
# Use formula 'Y ~ X' to indicate that we want to predict Y based on X
model = smf.ols('Y ~ X', data=data).fit()

# Print the summary of the regression results
print(model.summary())

# Plot the data points and the fitted regression line
plt.scatter(X, Y, color='blue', alpha=0.6, label="Data points")  # Data points
plt.plot(X, model.fittedvalues, color='red', label="Fitted regression line", linestyle='--')  # Fitted line
plt.title("Fitted Simple Linear Regression Model")
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.legend()
plt.show()
```

## Question Three
These two lines differ because of the error variable. The unfitted line includes the error data while the fitted model's line does not take the error data into account, rather just the slope of the correlation between the x variable and y.

```python
plt.plot(X, model.fittedvalues, color='red', label="Fitted regression line", linestyle='--')
plt.plot(X, beta_0 + beta_1 * X, color='green', label="True regression line", linestyle='--')
```
